```python
import numpy as np
import matplotlib.pyplot as plt
from tqdm import tqdm
from celluloid import Camera

plt.style.use('classic')
plt.rcParams["font.family"] = 'sans-serif'
```

```python
def vlcty(x,y):
    
    u1 = np.sin(x)*np.cos(y)
    
    u2 = -np.cos(x)*np.sin(y)
    
    return [u1, u2]

def init_posit(part):
    
    global x_p, y_p
    
    x_p=np.empty(shape=(qtt_step,1),
                 dtype=np.float32)
    
    y_p=np.empty(shape=(qtt_step,1),
                 dtype=np.float32)
    
    x_p[0] = x1[part]
    y_p[0] = x2[part]
    

def forward_euler(part=0,situation='init',list_iter=[1,1]):
    
    if situation=='init':
        for count in range(list_iter[0],list_iter[0]+1,1):
            x_p[count] = x_p[count-1] + vlcty(x_p[count-1],y_p[count-1])[0]*dt
            y_p[count] = y_p[count-1] + vlcty(x_p[count-1],y_p[count-1])[1]*dt
            
    elif situation=='global':
        for count in range(1,qtt_step,1):
            x_p[count] = x_p[count-1] + vlcty(x_p[count-1],y_p[count-1])[0]*dt
            y_p[count] = y_p[count-1] + vlcty(x_p[count-1],y_p[count-1])[1]*dt
            
        armzn_pos[f'EUL_{part}']=[x_p.copy(),y_p.copy()]
        
    
def AB2(part=0,situation='init',list_iter=[2,2]):
    
    if situation=='init':
        for count in range(list_iter[0],list_iter[0]+1,1):
            
            x_p[count] = x_p[count-1] + (dt/2)*(+3*vlcty(x_p[count-1],y_p[count-1])[0]
                                                -1*vlcty(x_p[count-2],y_p[count-2])[0])
 
            y_p[count] = y_p[count-1] + (dt/2)*(+3*vlcty(x_p[count-1],y_p[count-1])[1]
                                                -1*vlcty(x_p[count-2],y_p[count-2])[1])
    
    elif situation=='global':
        
        forward_euler()
        
        for count in range(2,qtt_step,1):
            x_p[count] = x_p[count-1] + (dt/2)*(+3*vlcty(x_p[count-1],y_p[count-1])[0]
                                                -1*vlcty(x_p[count-2],y_p[count-2])[0])
 
            y_p[count] = y_p[count-1] + (dt/2)*(+3*vlcty(x_p[count-1],y_p[count-1])[1]
                                                -1*vlcty(x_p[count-2],y_p[count-2])[1])
    
        armzn_pos[f'AB2_{part}']=[x_p.copy(),y_p.copy()]
    
def AB3(part=0,situation='init',list_iter=[3,3]):
    
    if situation=='init':
        for count in range(list_iter[0],list_iter[0]+1,1):
    
            x_p[count] = x_p[count-1] + (dt/12)*(+23*vlcty(x_p[count-1],y_p[count-1])[0]
                                                 -16*vlcty(x_p[count-2],y_p[count-2])[0]
                                                 + 5*vlcty(x_p[count-3],y_p[count-3])[0])

            y_p[count] = y_p[count-1] + (dt/12)*(+23*vlcty(x_p[count-1],y_p[count-1])[1]
                                                 -16*vlcty(x_p[count-2],y_p[count-2])[1]
                                                 + 5*vlcty(x_p[count-3],y_p[count-3])[1])

    
    elif situation=='global':
        
        forward_euler()
        AB2()
        
        for count in range(3,qtt_step,1):
            x_p[count] = x_p[count-1] + (dt/12)*(+23*vlcty(x_p[count-1],y_p[count-1])[0]
                                                 -16*vlcty(x_p[count-2],y_p[count-2])[0]
                                                 + 5*vlcty(x_p[count-3],y_p[count-3])[0])
    
            y_p[count] = y_p[count-1] + (dt/12)*(+23*vlcty(x_p[count-1],y_p[count-1])[1]
                                                 -16*vlcty(x_p[count-2],y_p[count-2])[1]
                                                 + 5*vlcty(x_p[count-3],y_p[count-3])[1])
        
        armzn_pos[f'AB3_{part}']=[x_p.copy(),y_p.copy()]
    
def AB4(part,situation='global'):
    
    forward_euler()
    AB2()
    AB3()
    
    for counter in range(4,qtt_step,1):
        
        x_p[counter] = x_p[counter-1] + (dt/24)*(+55*vlcty(x_p[counter-1],y_p[counter-1])[0]
                                                 -59*vlcty(x_p[counter-2],y_p[counter-2])[0]
                                                 +37*vlcty(x_p[counter-3],y_p[counter-3])[0]
                                                 - 9*vlcty(x_p[counter-4],y_p[counter-4])[0])
        
        y_p[counter] = y_p[counter-1] + (dt/24)*(+55*vlcty(x_p[counter-1],y_p[counter-1])[1]
                                                 -59*vlcty(x_p[counter-2],y_p[counter-2])[1]
                                                 +37*vlcty(x_p[counter-3],y_p[counter-3])[1]
                                                 - 9*vlcty(x_p[counter-4],y_p[counter-4])[1])
        
    
    armzn_pos[f'AB4_{part}']=[x_p.copy(),y_p.copy()]
        

def particle_color(part):
    
    global prtc_color
    
    prtc_color={}
    
    for count in range(0,part,1):
        prtc_color[f'{count}']=[tuple(np.random.uniform(low=0,high=1,size=3)) +(1.0,)]
                                                        
```

```python
armzn_pos={}

tn=32

t0=0

qtt_step=75

n_particles=1

x1=np.random.uniform(0,np.pi,size=n_particles)

x2=np.random.uniform(0,np.pi,size=n_particles)

dt=(tn-t0)/qtt_step

dt_plot = "{:.2f}".format(dt)
```

```python
fig,ax = plt.subplots(nrows=2,ncols=2,figsize = (12, 12), sharex=True, sharey=True)

camera = Camera(fig)

plt.xlim(-0.5,2*np.pi+0.5), plt.ylim(-0.5,2*np.pi+0.5)

ax[0,0].set_title(f'AB4'), ax[0,1].set_title(f'AB3'), ax[1,0].set_title(f'AB2'), ax[1,1].set_title(f'FEU')

particle_color(n_particles)

for loop in tqdm(range(1,qtt_step+1,1)):
    
    for particles in range(0,n_particles,1):

        init_posit(part=particles)
        AB4(part=particles, situation='global')
        ax[0,0].plot(armzn_pos[f'AB4_{particles}'][0],armzn_pos[f'AB4_{particles}'][1], alpha=0.15, linestyle='--', color=prtc_color[f'{particles}'][0])
        ax[0,0].plot(armzn_pos[f'AB4_{particles}'][0][0:loop],armzn_pos[f'AB4_{particles}'][1][0:loop], alpha=1, color=prtc_color[f'{particles}'][0])

        init_posit(part=particles)
        AB3(part=particles, situation='global')
        ax[0,1].plot(armzn_pos[f'AB3_{particles}'][0],armzn_pos[f'AB3_{particles}'][1], alpha=0.15, linestyle='--', color=prtc_color[f'{particles}'][0])
        ax[0,1].plot(armzn_pos[f'AB3_{particles}'][0][0:loop],armzn_pos[f'AB3_{particles}'][1][0:loop], alpha=1, color=prtc_color[f'{particles}'][0])

        init_posit(part=particles)
        AB2(part=particles, situation='global')
        ax[1,0].plot(armzn_pos[f'AB2_{particles}'][0],armzn_pos[f'AB2_{particles}'][1], alpha=0.15, linestyle='--', color=prtc_color[f'{particles}'][0])
        ax[1,0].plot(armzn_pos[f'AB2_{particles}'][0][0:loop],armzn_pos[f'AB2_{particles}'][1][0:loop], alpha=1, color=prtc_color[f'{particles}'][0])

        init_posit(part=particles)
        forward_euler(part=particles, situation='global')
        ax[1,1].plot(armzn_pos[f'EUL_{particles}'][0],armzn_pos[f'EUL_{particles}'][1], alpha=0.15, linestyle='--', color=prtc_color[f'{particles}'][0])
        ax[1,1].plot(armzn_pos[f'EUL_{particles}'][0][0:loop],armzn_pos[f'EUL_{particles}'][1][0:loop], alpha=1, color=prtc_color[f'{particles}'][0])
        
    plt.suptitle(f'LPT DIFFERENT METHODS \n dt={dt_plot}, n_iter={qtt_step}, {n_particles} particles', fontsize=16.5)       
        
    camera.snap()
    
fig.tight_layout(rect=[0, 0.03, 1, 0.95])

animation = camera.animate(interval=int(0.3*60*1000/qtt_step))

animation.save('testing_8.mp4')
```
